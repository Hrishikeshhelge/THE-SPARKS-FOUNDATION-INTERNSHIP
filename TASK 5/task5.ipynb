Import essential libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('matches.csv')
df.head()
```

Visualization

```python
sns.countplot(x='season', data=df)
```

Match Winning Teams

```python
sns.countplot(x='winner', data=df)
plt.title('Match Winning Teams', fontsize=20)
plt.xticks(rotation=90)
```

Indian Premier League Venue

```python
plt.subplots(figsize=(20,10))
sns.countplot(x='venue',data=df,order=df['venue'].value_counts().index)
plt.title('Indian Premier League Venue', fontsize=30)
plt.xlabel('Stadium')
plt.xlabel('Matches')
plt.xticks(rotation=-90)
```

Top player of the amtch winners

```python
plt.figure(figsize=(20,10))

sns.barplot(x= df.player_of_match.value_counts()[:10].index,y=df.player_of_match.value_counts()[:10])

plt.title('Top player of the amtch winners', fontsize=30)
plt.xlabel('Player')
plt.xlabel('Match')
```

Champion Teams

```python
plt.subplots(figsize=(20,10))
sns.countplot('winner', data=df)
plt.title('Champion Teams', fontsize=25)
plt.xlabel('Team')
plt.xlabel('Season won')
plt.xticks(rotation=-90)
plt.show()
```

### Conclusion
* Mumbai indains plays the highest amount of matches and has won more seasons than the other teams
* Chennai Super Kings ranked the second highest in winning matches followed by Kolkata Knight Riders
* Eden Gardens hosts the highest number of matches followed by M Chinnaswamy Stadium and wankhede Stadium
* CH Gayle won the most MVP awards in Indian Premier League
